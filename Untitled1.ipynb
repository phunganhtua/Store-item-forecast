```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv('train.csv',parse_dates=['date'])
df_test = pd.read_csv('test.csv',parse_dates=['date'])
df_test = df_test.drop(['id'],axis=1)
```

```python
#Data by month over year
from datetime import datetime
df1 = df_train[(df_train['store']==1)&(df_train['item']==1)].reset_index()
df1['date'] = df1['date'].dt.strftime('%Y-%m')
df1 = df1.groupby('date')['sales'].sum().reset_index()
df1.date = pd.to_datetime(df1.date)
df1 = df1.set_index('date')
df1
```

```python
df1.plot()
```

```python
df1_stat = df1.diff().dropna()
df1_stat
```

```python
#Check for stationary
from statsmodels.tsa.stattools import adfuller
res = adfuller(df1_stat)
print('ADF Statistic: %f' % res[0])
print('p-value: %f' % res[1])
```

```python
import statsmodels.tsa.api as smt
smt.graphics.plot_acf(df1_stat, lags=36, color='mediumblue')
smt.graphics.plot_pacf(df1_stat, lags=24, color='mediumblue')
```

```python
pp = range(0,7)
d = 1
qq = range(0,7)
P=1
D=0
Q=0
```

```python
#Train test split 
df1_train = df1[:-12]
df1_test = df1[-12:]
df1_train.head()
```

```python
import statsmodels.api as sm
import warnings 
warnings.simplefilter('ignore')
scores = []
for p in pp: 
    for q in qq:
        if p==0 and q==0: 
            continue
        try:
            mod = sm.tsa.statespace.SARIMAX(df1_train, order=(p,d,q), seasonal_order=(P,D,Q,12))
            res = mod.fit(disp=False)
            score = [p,d,q,P,D,Q,res.aic,res.aicc,res.bic,res.llf]
            scores.append(score)
            del mod
            del res
        except:
            pass
                        
res = pd.DataFrame(scores)

res.columns = ['p', 'd', 'q', 'P', 'D', 'Q','AIC','AICC','BIC','LLF']
```

```python
models = res.sort_values(['AICC','AIC','LLF','BIC'],ascending=[True,True,False,True]).head(10)
models
```

```python
models=models.reset_index(drop=True)
Order_List=[]
for y in range(10):
    o=(int(models.iloc[y][0]),int(models.iloc[y][1]),int(models.iloc[y][2]))
    so=(int(models.iloc[y][3]),int(models.iloc[y][4]),int(models.iloc[y][5]),12)
    OL=[o,so]
    Order_List.append(OL)
Order_List
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from math import sqrt
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / np.maximum(np.ones(len(y_true)), np.abs(y_true))))*100
```

```python
def plot_sarima (dataframe, order, seasonal_order,steps):
    mod = sm.tsa.statespace.SARIMAX(dataframe, order=order, seasonal_order=seasonal_order)
    res = mod.fit(disp=False)
    predict=res.forecast(steps)
    
    #Quality Measures
    MSE= mean_squared_error(df1_test,predict)
    RMSE=format(sqrt(MSE),'.2f')
    MAE= format(mean_absolute_error(df1_test,predict),'.2f')
    MAPE=format(mean_absolute_percentage_error(df1_test,predict),'.2f')
    R2= format(r2_score(df1_test,predict),'.2f')
    qm=[order,seasonal_order,RMSE,MAE,MAPE,R2]
    
    # Plot
    fig, ax = plt.subplots(figsize=(18,6))
    ax.xaxis.grid(2)
    ax.plot(dataframe,'black', label="train")
    ax.plot(df1_test , 'red', label= 'test'),
    ax.plot( predict, 'b--', linestyle='--',linewidth=2, label='forcast')

    ax.set(title= 'Order: {0}{1}; RMSE={2}; MAE={3}; MAPE={4}; R2={5} '.format(order,seasonal_order, RMSE,MAE,MAPE,R2));
    ax.legend(loc="upper left")
    return qm
```

```python
QM=[]
for x in Order_List:
    ##print(x[0])
    qm=plot_sarima(dataframe=df1_train, order=x[0], seasonal_order=x[1],steps=12)
    QM.append(qm)
Quality_Metrics=pd.DataFrame(QM)
```

```python
Quality_Metrics.columns = ['Order','Seasonal_Order','RMSE','MAE','MAPE','R2']
Quality_Metrics.sort_values(['R2','RMSE','MAPE', 'MAE'],ascending=[False,True,True,True],inplace=True)
Quality_Metrics
```

```python
def Res_Diag(order, seasonal_order):
    mod = sm.tsa.statespace.SARIMAX(df1, order=order, seasonal_order=seasonal_order)
    res = mod.fit(disp=False)
    df_resid=res.resid
    df_resid.plot()

    #ACF & PACF plot
    fig, ax = plt.subplots(2,1)
    fig =smt.graphics.plot_acf(df_resid, lags=13, ax=ax[0])
    fig =smt.graphics.plot_pacf(df_resid, lags=13, ax=ax[1])
    plt.show()

    #Histogram
    ax = pd.DataFrame(df_resid).plot(kind='hist')
    ax.set_ylabel('Number of occurences')
    ax.set_xlabel('Residual')
    plt.show()
    
    res.plot_diagnostics(figsize=(15, 12))
    plt.show()


    #Summary
    print(res.summary())
```

```python
order_1=(1,1,0)
seasonal_order = (1,0,0,12)
Res_Diag(order=order_1, seasonal_order=seasonal_order)
```

```python
order_4=(0,1,1)
seasonal_order_4=(0,1,0,12)
steps=11
mod_4 = sm.tsa.statespace.SARIMAX(df1, order=order_4, seasonal_order=seasonal_order_4)
res_4 = mod_4.fit(disp=False)
predict_4 = res_4.get_prediction(end=mod_4.nobs-1+steps)
```

```python
#90% confidence band
predict_ci_4 = predict_4.conf_int(alpha=.05)[-(steps+12):] 

# Graph
fig, ax = plt.subplots(figsize=(18,8))
#ax.xaxis.grid('both')
ax.plot( df1,'black',label='Item 1',marker='o',markersize=3)

# Plot
ax.plot( predict_4.predicted_mean[-(steps+12):-steps], 'red', label= 'In sample forcast',marker='o',markersize=3),
ax.plot( predict_4.predicted_mean[-(steps+1):], 'r--', label='Out sample forcast',marker='o',markersize=3)
ax.fill_between(predict_ci_4.index, predict_ci_4.iloc[:, 0], predict_ci_4.iloc[:, 1], color='blue', alpha=.1,label='95% confidence band')
ax.set(title='Forecast by Model 4:SARIMA{0}{1}'.format(order_4,seasonal_order_4))
ax.legend(loc="upper left")
```
