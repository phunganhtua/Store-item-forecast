```python
sales_base = df_train.groupby(['date', 'store', 'item'])['sales'].sum().reset_index()
sales_base.set_index('date', inplace=True)
# Create a pivot table to make each store and family pair a column
sales_base.reset_index(inplace=True)
sales_pivot = sales_base.pivot(index='date', columns=['store','item'], values='sales')
# Create a new index to have all the days in the period
new_index = pd.date_range(sales_pivot.index.min(), sales_pivot.index.max())
sales = sales_pivot.reindex(new_index, fill_value=0)
new_index = pd.date_range(sales_pivot.index.min(), sales_pivot.index.max())
sales = sales_pivot.reindex(new_index, fill_value=0)
sales
```

```python
sales.loc[:'2017-01-01'][(1,1)].info()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import itertools
from sklearn.metrics import mean_absolute_error
```

```python
alphas = betas = gammas = np.arange(0.20, 1, 0.05)
abg = list(itertools.product(alphas, betas, gammas))
```

```python
train = sales.loc[:'2016-12-31'][(1,1)]
val = sales.loc['2017-01-01':][(1,1)]
import warnings
warnings.filterwarnings("ignore")
def tes_optimizer(train, abg, step=365):
    best_alpha, best_beta, best_gamma, best_mae = None, None, None, float("inf")
    for comb in abg:
        tes_model = ExponentialSmoothing(train, trend="add", seasonal="add", seasonal_periods=7).\
            fit(smoothing_level=comb[0], smoothing_trend=comb[1], smoothing_seasonal=comb[2])
        y_pred = tes_model.forecast(step)
        mae = sum((y_pred-val.values)**2)/365
        if mae < best_mae:
            best_alpha, best_beta, best_gamma, best_mae = comb[0], comb[1], comb[2], mae
    print("best_alpha:", round(best_alpha, 2), "best_beta:", round(best_beta, 2), "best_gamma:", round(best_gamma, 2),
          "best_mae:", round(best_mae, 4))

    return best_alpha, best_beta, best_gamma, best_mae
best_alpha, best_beta, best_gamma, best_mae = tes_optimizer(train, abg,step=365)
```

```python
model = ExponentialSmoothing(train,trend='add',seasonal='add',seasonal_periods=7).fit(smoothing_level=0.6,smoothing_trend=0.2,smoothing_seasonal=0.25)
y_pred = model.forecast(365+92)
```
